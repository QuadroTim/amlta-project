```python
from amlta.config import config
from amlta.question_generation.process import (
    build_training_batched_dfs_all,
)
```

```python
tapas_train_batched_dfs = build_training_batched_dfs_all(shuffle_rows=False)
tapas_train_batched_dfs
```

```python
tapas_train_batched_dfs["question_id"] = tapas_train_batched_dfs["question_id"].astype(
    "int"
)

path = config.data_dir / "tapas-ft" / "data" / "tapas_train_batched_dfs.parquet"
path.parent.mkdir(parents=True, exist_ok=True)
tapas_train_batched_dfs.to_parquet(path)
```

```python
tapas_train_batched_dfs_shuffled = build_training_batched_dfs_all(shuffle_rows=True)
tapas_train_batched_dfs_shuffled
```

```python
tapas_train_batched_dfs_shuffled["question_id"] = tapas_train_batched_dfs_shuffled[
    "question_id"
].astype("int")

path = (
    config.data_dir / "tapas-ft" / "data" / "tapas_train_batched_dfs_shuffled.parquet"
)
path.parent.mkdir(parents=True, exist_ok=True)
tapas_train_batched_dfs_shuffled.to_parquet(path)
```
